```python
# this function will tell the user if their grades are good to be a game developer
#this is a function which works wih the old grading system.

def askgcse():

    #global allows variables to be used in a different function
    global math
    global comp
    global point

    #this will be the variable to store the points to determine the final outcome
    point = 0
    gcse = input("Did you do the old gcse or the new ones? (old/new) : ")

    gcse = gcse.upper()

    # if the gcse is the old grading system it will capitalise and check each grade
    if gcse == "OLD":
        math = input("What grade did you get in your math GCSE's (F - A) : ")
        comp = input("What grade did you get in your computing (F - A) : ")
        math = math.upper()
        comp = comp.upper()
        old()

    # if the grading system is new it will only capitalise the computing gcse
    # because math grade is a number.
    elif gcse == "NEW":
        math = input("What grade did you get in your math GCSE's (1 - 9) : ")
        comp = input("What grade did you get in yourcomputing (F - A) : ")
        comp = comp.upper()
        new()

    else:
        print("Sorry but other qalifications are not allowed. ")
        askgcse()
        
#this is a different function which allows the user to input the old grading system
def old():
    
    if math == "B" or comp == "B":
        print("ok")
        
    elif math == "A" or comp == "A":
        print("ok")

    else:
        print ("You are unlikely to be able to continue in this career path.")
        exit()
        


#this function will work with the new math grades
#it will check if the math grades are B(6) or above the same a computing grades
def new():
    
    if math == "6" or comp == "B":
        print ("ok")

    elif math == "7" or comp == "A":
        print ("ok")
        
    elif math == "8" or math == "9":
        print ("ok")
        
    else:
        print ("You are unlikely to be able to continue in this career path.")
        exit()
        
#this function will ask the user what they plan to do next
def part_two():
    
    global btec_alevel
    global point

    btec_alevel = input ("Did you choose A-level(math) or BTEC (computing) or Both? (BTEC/ALEVEL/BOTH/NONE) :")

    btec_alevel = btec_alevel.upper()

    if btec_alevel == "BTEC":
        point = point + 60

    elif btec_alevel == "ALEVEL":
        point = point + 60

    elif btec_alevel == "BOTH":
        point = point + 100
        
    elif btec_alevel == "NONE":
        print ("You are unlikely to be able to continue in this career path.")
        exit()

    #if the answer given is different to BTEC/ALEVEL/BOTH/NONE it will say that the answer isn't accepted  and will repeat the question
    else:
        print ("Sorry That is not a valid answer : ")

        part_two()
    
#Pass grade give them another 100 points, for Merit grade give another 150 points, for Distinction another 200, for Distinction* another 300
def btec_grade():

    global grade
    global point

    if btec_alevel == "BTEC" or btec_alevel == "BOTH" :

        grade = input ("What grade do you think you will get in your btec course? (PASS/MERIT/DISTINCTION/DISTINCTION*): ")

        grade = grade.upper()
        
        if grade == "PASS":
            point = point + 100

        elif grade == "MERIT":
            point = point + 150

        elif grade == "DISTINCTION":
            point = point + 200

        elif grade == "DISTINCTION*":
            point = point + 300
        
        else:
            print ("That is not a valid answer.")

            btec_grade()

    #print (point)
       
def work():
    
    global games
    global work
    global point
    
    games = input("Have you made any games? (YES/NO) : ")

    games = games.upper()

    if games == "YES":

        point = point + 500
        
    #print(point)

    work = input("Do you have any work experience in the industry? (YES/NO) : ")

    work = work.upper()

    if work == "YES":
        point = point +300

    #print(point)


def calculation():
    
    global point

    if point <= 300:
        print("You are somewhat unlikely to be able to continue in this career path")

    elif point > 300 and point <= 1000:
        print ("You may be able to continue in this career path, but you may find regular work difficult to find")
    
    elif point > 1000:
        print ("You have a chance in this career path, but be aware it is very competitive")
    


#runs all the functions in order which will allow the input to effect the points and will not cause any errors.
askgcse()
part_two()
btec_grade()
work()
calculation()
exit()
#print (point)
```
